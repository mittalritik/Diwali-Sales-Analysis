# Exploratory Data Analysis

Gender

```python
# Create the count plot
ax = sns.countplot(x="Gender", data=df, palette=["cyan","orange"])
ax.set_title("Gender Count Plot")

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_gen = df.groupby(["Gender"], as_index= False)["Amount"].sum().sort_values(by="Amount", ascending= False)
ax = sns.barplot(x = "Gender", y = "Amount", data = sales_gen, palette=["red","green"])
ax.set_title("Gender Amount Plot")
plt.show()
```

Age

```python
ax = sns.countplot(data = df, x = "Age Group", palette= ["yellow", "orange"], hue= "Gender")
ax.set_title("Age Count Plot")

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)["Amount"].sum().sort_values(by= "Amount", ascending= False)
ax = sns.barplot(x= "Age Group", y= "Amount", data= sales_age)
sns.set_palette('coolwarm')
ax.set_title("Age Amount Plot")

plt.show()
```

State

```python
orders_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending = False).head(8)
ax = sns.barplot(x = 'State', y = 'Orders', data = orders_state)
sns.set_palette('Dark2')  
ax.set_title('Statewise Orders')
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(8)
ax = sns.barplot(x = 'State', y = 'Amount', data = sales_state, palette=['red','green','yellow','orange','cyan']) 
ax.set_title('Statewise Sales')
plt.show()
```

From the above graph we can see that most of the orders and total sales/ amount are from Uttar Pradesh, Maharashtra, and Karnataka.

Marital Status

```python
sns.set(rc={'figure.figsize':(5,5)})
ax = sns.countplot(data = df , x = 'Marital_Status' , palette = ['yellow','pink'])

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_m_s = df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax = sns.barplot(data = sales_m_s , x = 'Marital_Status' , y = 'Amount' , hue = 'Gender')
ax.set_title('Marital_Status Sales Plot')
plt.show()
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data= df, x='Occupation')
ax.set_title('Occupation Count Plot')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_occupation = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax = sns.barplot(data = sales_occupation , x = 'Occupation' , y = 'Amount' , palette = ['cyan'])

ax.set_title('Occupation Sales Plot')
plt.show()
```

Product Category

```python
ax = sns.countplot(data= df, x='Product_Category')
ax.set_title('Product_Category Count Plot')
plt.xticks(rotation= 45)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_category = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)
ax = sns.barplot(data = sales_category , x = 'Product_Category' , y = 'Amount' , palette = ['cyan'])
plt.xticks(rotation=45)

ax.set_title('Product_Category Sales Plot')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
count_product_ID = df.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)
ax = sns.barplot(data = count_product_ID , x = 'Product_ID' , y = 'Orders')
plt.xticks(rotation=0)

ax.set_title('Product_ID Count Plot')
plt.show()
```
